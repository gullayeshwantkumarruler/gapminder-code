## Gapminder_gdp_asia

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data_asia = pd.read_excel('/content/gapminder_gdp_asia (5).xlsx')

data_asia.head()
```

```python
chin_ind=data_asia[ (data_asia.country=='India') | (data_asia.country=='China') ]
chin_ind.head()
```

```python
chin_ind=chin_ind.set_index('country')
```

```python
# chin_ind.plot(kind='line',figsize=(20,10),title="GDP Per Capital plot",xlabel="gdp_Per_cap_Per_Year")
```

```python
# code to plot the per capita values
chin_ind.T.plot(kind='line',figsize=(20,10),title="GDP Per Capital plot",xlabel="gdp_Per_cap_Year",ylabel='GDP per capita')
```
